## Resolución

### Obtener la función transferencia

Planteo las ecuaciones:

(1) $\frac{(V_X-V_1)}{R_1} + \frac{V_X-V_2}{R_2} = 0$

(2) $\frac{V_X}{R_3} + (V_X-V_1) . SC = 0$

Partiendo de la ecuación (1):

$V_X . (\frac{1}{R_1} + \frac{1}{R_2}) - \frac{V_1}{R_1} - \frac{V_2}{R_2} = 0$

$V_X . \frac{R_1+R_2}{R_1 . R_2} - \frac{V_1}{R_1} = \frac{V_2}{R_2}$ (3)

Partiendo de la ecuación (2):

$V_X . (\frac{1}{R_3} + SC) - V_1 . SC = 0$

$V_X = \frac{V_1 . SC}{\frac{1}{R_3}+SC}$ (4)

Reemplazando (4) en (3):

$V_1 . \frac{SC}{\frac{1}{R_3}+SC} \frac{R_1+R_2}{R_1 . R_2} - \frac{V_1}{R_1} = \frac{V_2}{R_2}$

$V_1 . (\frac{SC}{\frac{1}{R_3}+SC} \frac{R_1+R_2}{R_1 . R_2} - \frac{1}{R_1}) = \frac{V_2}{R_2}$

$V_1 . (\frac{SR_1^2C+SR_1R_2C-SR_1R_2C-\frac{R_1R_2}{R_3}}{SR_1^2R_2C+\frac{R_1^2R_2}{R_3}}) = \frac{V_2}{R_2}$

$V_1 . \frac{S-\frac{R_2}{R_1R_3C}}{S+\frac{1}{R_3C}} = V_2$

$T(S) = \frac{V_2}{V_1} = \frac{S-\frac{R_2}{R_1R_3C}}{S+\frac{1}{R_3C}}$

### Normalización de la función transferencia

Definimos $\omega_0 = \frac{1}{R_3C}$

Suponemos el caso $R_1=R_2$

Por lo tanto, la transferencia quedaría:

$T(S) = \frac{S-\omega_0}{S+\omega_0}$

Normalizo en frecuencia: $S = S'\omega_0$

$T(S') = \frac{S'\omega_0-\omega_0}{S'\omega_0+\omega_0}$

$T(S') = \frac{S'-1}{S'+1}$

```python
# Simulación de la Transferencia Normalizada

import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig
from splane import analyze_sys, pretty_print_bicuad_omegayq

# Señal normalizada
num = np.array([1,-1])
den = np.array([1,1])

pretty_print_bicuad_omegayq(num, den)
transfer_f = sig.TransferFunction(num, den)

plt.close('all')
analyze_sys(transfer_f, 'Transferencia')
```
