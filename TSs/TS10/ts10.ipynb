## Ejercicio 1 - Resolución

Para sintetizar $Z(s)$ en derivación, se debe considerar $Y(s) = \frac{1}{Z(s)}$:

$Y(s) = \frac{s.(s^2+2)}{(s^2+3).(s^2+1)}$

A) Síntesis por el método Foster

$Y(s) = \frac{K_0}{s} + \frac{2.K_1.s}{s^2+1} + \frac{2.K_2.s}{s^2+3} + K_{\infty} . s$

$K_0 = \lim_{s \to 0} s.Y(s) = \lim_{s \to 0} s.\frac{s.(s^2+2)}{(s^2+3).(s^2+1)} = \frac{0}{3} = 0$

$2.K_1 = \lim_{s \to j1} \frac{s^2+1}{s} . Y(s) = \lim_{s \to j1} \frac{s^2+1}{s} . \frac{s.(s^2+2)}{(s^2+3).(s^2+1)} = \frac{-1+2}{-1+3} = \frac{1}{2}$

$2.K_2 = \lim_{s \to j\sqrt{3}} \frac{s^2+3}{s} . Y(s) = \lim_{s \to j\sqrt{3}} \frac{s^2+3}{s} . \frac{s.(s^2+2)}{(s^2+3).(s^2+1)} = \frac{-3+2}{-3+1} = \frac{1}{2}$

$K_{\infty} = \lim_{s \to \infty} \frac{1}{s} . Y(s) = \lim_{s \to \infty} \frac{1}{s} . \frac{s.(s^2+2)}{(s^2+3).(s^2+1)} = \lim_{s \to \infty} \frac{s^2.(1+\frac{2}{s^2})}{s^4.(1+\frac{3}{s^2}).(1+\frac{1}{s^2})} = 0$

Finalmente:

$Y(s) = \frac{\frac{1}{2}.s}{s^2+1} + \frac{\frac{1}{2}.s}{s^2+3} = \frac{1}{2.s+\frac{1}{\frac{1}{2}.s}} + \frac{1}{2.s+\frac{1}{\frac{1}{6}.s}}$

B)

Síntesis por el método de Cauer 1 (remoción de polos en alta frecuencia)

$Z(s) = \frac{(s^2+3).(s^2+1)}{s.(s^2+2)}$

<div><img src="img/ej1_b_cauer1_1.jpeg" width="500"/></div>

$Z_2(s) = \frac{(s^2+3).(s^2+1)}{s.(s^2+2)} - K_{\infty}.s = \frac{s^4+4.s^2+3-K_{\infty}.s^2.(s^2+2)}{s.(s^2+2)}$

$Z_2(s) = \frac{s^4+4.s^2+3-K_{\infty}.s^4-2.K_{\infty}.s^2}{s.(s^2+2)}$

$K_{\infty} = \lim_{s \to \infty} \frac{Z(s)}{s} = \lim_{s \to \infty} \frac{(s^2+3).(s^2+1)}{s^2.(s^2+2)} = 1$

**Esto representa un inductor en serie de valor:** $K_{\infty} = 1$

$Z_2(s) = \frac{4.s^2+3-2.s^2}{s.(s^2+2)} = \frac{2.s^2+3}{s.(s^2+2)}$

Invertimos $Z_2(s)$:

$Y_2(s) = \frac{1}{Z_2(s)} = \frac{s.(s^2+2)}{2.(s^2+\frac{3}{2})}$

<div><img src="img/ej1_b_cauer1_2.jpeg" width="500"/></div>

$Y_4(s) = \frac{s.(s^2+2)}{2.(s^2+\frac{3}{2})} - K_{2\infty}.s = \frac{s^3+2.s-K_{2\infty}.s.(2.(s^2+\frac{3}{2}))}{2.(s^2+\frac{3}{2})}$

$K_{2\infty} = \lim_{s \to \infty} \frac{Y_2(s)}{s} = \lim_{s \to \infty} \frac{s^2+2}{2.(s^2+\frac{3}{2})} = \frac{1}{2}$

**Esto representa un capacitor en derivación de valor:** $K_{2\infty} = \frac{1}{2}$

$Y_4(s) = \frac{\frac{1}{2}.s}{2.(s^2+\frac{3}{2})}$

Invertimos $Y_4(s)$:

$Z_4(s) = \frac{4.(s^2+\frac{3}{2})}{s}$

<div><img src="img/ej1_b_cauer1_3.jpeg" width="500"/></div>

$Z_6(s) = \frac{4.(s^2+\frac{3}{2})}{s} - K_{3\infty}.s = \frac{4.s^2+6-K_{3\infty}.s^2}{s}$

$K_{3\infty} = \lim_{s \to \infty} \frac{Z_4(s)}{s} = \lim_{s \to \infty} \frac{1}{s} . \frac{4.(s^2+\frac{3}{2})}{s} = 4$

**Esto representa un inductor en serie de valor:** $K_{3\infty} = 4$

$Z_6(s) = \frac{4.s^2+6-4.s^2}{s} = \frac{6}{s} = \frac{1}{\frac{1}{6}.s}$

**Esto representa un capacitor en derivación de valor:** $\frac{1}{6}$

Síntesis por el método de Cauer 2 (remoción de polos en baja frecuencia)

$Z(s) = \frac{(s^2+3).(s^2+1)}{s.(s^2+2)}$

<div><img src="img/ej1_b_cauer2_1.jpeg" width="500"/></div>

$Z_2(s) = \frac{(s^2+3).(s^2+1)}{s.(s^2+2)} - \frac{K_0}{s} = $

$K_0 = \lim_{s \to 0} s.Z(s) = \lim_{s \to 0} s.\frac{(s^2+3).(s^2+1)}{s.(s^2+2)} = \frac{(0+3).(0+1)}{0+2} = \frac{3}{2}$

**Esto representa un capacitor en serie de valor:** $\frac{1}{K_0} = \frac{2}{3}$

$Z_2(s) = \frac{(s^2+3).(s^2+1)- \frac{K_0}{s}.s.(s^2+2)}{s.(s^2+2)} = \frac{s^4+4.s^2+3-\frac{3}{2}.s^2-3}{s.(s^2+2)} = \frac{s^4+\frac{5}{2}.s^2}{s.(s^2+2)}$

Invertimos $Z_2(s)$:

$Y_2(s) = \frac{s.(s^2+2)}{s^4+\frac{5}{2}.s^2}$

<div><img src="img/ej1_b_cauer2_2.jpeg" width="500"/></div>

$Y_4(s) = \frac{s.(s^2+2)}{s^4+\frac{5}{2}.s^2} - \frac{K_{01}}{s} = \frac{s^3+2.s-\frac{K_{01}}{s}.(s^4+\frac{5}{2}.s^2)}{s^4+\frac{5}{2}.s^2}$

$K_{01} = \lim_{s \to 0} s.Y_2(s) = \lim_{s \to 0} \frac{s^2.(s^2+2)}{s^4+\frac{5}{2}.s^2} = \lim_{s \to 0} \frac{s^2.(s^2+2)}{s^2.(s^2+\frac{5}{2})} = \frac{0+2}{0+\frac{5}{2}} = \frac{4}{5}$

**Esto representa un inductor en derivación de valor:** $\frac{1}{K_{01}} = \frac{5}{4}$

$Y_4(s) = \frac{\frac{1}{5}.s^3}{s^4+\frac{5}{2}.s^2}$

Invertimos $Y_4(s)$:

$Z_4(s) = \frac{s^4+\frac{5}{2}.s^2}{\frac{1}{5}.s^3} = \frac{s^2+\frac{5}{2}}{\frac{1}{5}.s}$

<div><img src="img/ej1_b_cauer2_3.jpeg" width="500"/></div>

$Z_6(s) = \frac{s^2+\frac{5}{2}}{\frac{1}{5}.s} - \frac{K_{02}}{s} = \frac{s^2+\frac{5}{2}-\frac{K_{02}}{s}.\frac{1}{5}.s}{\frac{1}{5}.s}$

$K_{02} = \lim_{s \to 0} Z_4(s) . s = \lim_{s \to 0} \frac{s^2+\frac{5}{2}}{\frac{1}{5}.s}.s = 5 . (0+\frac{5}{2}) = \frac{25}{2}$

**Esto representa un capacitor en serie de valor:** $\frac{1}{K_{02}} = \frac{2}{25}$

$Z_6(s) = \frac{s^2+\frac{5}{2}-\frac{25}{2.s}.\frac{1}{5}.s}{\frac{1}{5}.s} = \frac{s^2+\frac{5}{2}-\frac{5}{2}}{\frac{1}{5}.s} = 5.s$

**Esto representa un inductor en derivación de valor:** 5

## Ejercicio 2 - Resolución

$Y(s) = \frac{3.s.(s^2+\frac{7}{3})}{(s^2+2).(s^2+5)}$

$Z(s) = \frac{1}{Y(s)} = \frac{(s^2+2).(s^2+5)}{3.s.(s^2+\frac{7}{3})}$

<div><img src="img/ej2_eje.jpeg" width="500"/></div>

Realizamos una remoción parcial del polo en continua para ubicar un cero en $\omega=1$

$K_0' = Z(s).s|_{s=j1}$

$K_0' = \frac{(s^2+2).(s^2+5)}{3.s.(s^2+\frac{7}{3})}.s|_{s=j1} = \frac{((-1)+2).((-1)+5)}{3.((-1)+\frac{7}{3})} = \frac{1.4}{(-3)+7}=1$

Por lo tanto $C_1 = 1$

$Z_2(s) = Z(s) - \frac{K_0'}{s} = \frac{(s^2+2).(s^2+5)}{3.s.(s^2+\frac{7}{3})}-\frac{1}{s} = \frac{s^4+7.s^2+10-3.s^2-7}{3.s.(s^2+\frac{7}{3})} = \frac{s^4+4.s^2+3}{3.s.(s^2+\frac{7}{3})}$

Invertimos $Z_2(s)$:

$Y_2(s) = \frac{3.s.(s^2+\frac{7}{3})}{s^4+4.s^2+3}$

$2.K_1' = Y_2(s).\frac{s^2+1}{s}|_{s=j1} = \frac{3.s.(s^2+\frac{7}{3})}{s^4+4.s^2+3} . \frac{s^2+1}{s}|_{s=j1} = \frac{3.s.(s^2+\frac{7}{3})}{(s^2+1).(s^2+3)}.\frac{s^2+1}{s}|_{s=j1} = \frac{3.(s^2+\frac{7}{3})}{s^2+3}|_{s=j1} = \frac{-3+7}{-1+3}$

$2.K_1' = 2$

$K_1' = 1$

Se remueve una rama LC serie que resuena en $j1$:

$C_2 = \frac{2.K_1'}{\omega_1^2} = 2$

$L_2 = \frac{1}{2.K_1'} = \frac{1}{2}$

Para la otra rama planteo:

$Y_4(s) = Y_2(s) - \frac{2.s}{s^2+1} = \frac{3.s.(s^2+\frac{7}{3})}{(s^2+1).(s^2+3)} - \frac{2.s}{s^2+1} = \frac{1}{s^2+1} . [\frac{3.s.(s^2+\frac{7}{3})}{s^2+3} - 2.s]$

$Y_4(s) = \frac{1}{s^2+1} . [\frac{3.s.(s^2+\frac{7}{3})- 2.s.(s^2+3)}{s^2+3}] = \frac{1}{s^2+1} . \frac{3.s^3+7.s-2.s^3-6.s}{s^2+3} = \frac{1}{s^2+1} . \frac{s^3+s}{s^2+3} = \frac{1}{s^2+1} . \frac{s.(s^2+1)}{s^2+3} = \frac{s}{s^2+3} = \frac{1}{\frac{s^2+3}{s}} = \frac{1}{1.s+\frac{1}{\frac{1}{3}.s}}$

$C_3 = \frac{1}{3}$

$L_3 = 1$
