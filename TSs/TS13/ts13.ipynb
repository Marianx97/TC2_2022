### Ejercicio 1

Sintetize la siguiente transferencia cargada con componentes RC:

![ej1_consigna](img/ej1_consigna.JPG)

$\frac{-I_2}{I_1}=H.\frac{s^2+5s+4}{s^2+8s+12}$

$Z_{21} = 6.H$

**a)** Obtener la topología mediante **síntesis gráfica**.

**b)** Calcular el valor de los componentes mediante **síntesis analítica**.


## Resolución

### a) Síntesis gráfica

Se puede considerar la carga $R_L$ como parte del cuadripolo y esto permite expresar la misma transferencia bajo una condición de carga de $V_2=0$:

<div><img src="img/cuadripolo_carga.jpeg" width="500"/></div>

De esta forma, se puede expresar que:

$\frac{(-I_2)}{I_1}|_{V_2=0} = \frac{-Y_{21}}{Y_{11}} = \frac{Z_{21}}{Z_{22}}$

Por lo tanto:

$\frac{6.H}{Z_{22}} = H.\frac{s^2+5s+4}{s^2+8s+12}$

Finalmente:

$Z_{22} = 6. \frac{s^2+8s+12}{s^2+5s+4}$

Verificando en los extermos de carga:

$Z_{22}(0) = 18$

$Z_{22}(\to \infty) = 6$

Se cumple que para $Z_{RC}$: $Z(0) > Z(\to \infty)$

El objetivo ahora es sintetizar $Z_{22}$ respetando los ceros de la transferencia, que son $(-1)$ y $(-4)$

Como se está sintetizando $Z_{22}$, se comienza a remover desde la salida. Lo primero en remover es un resistor en serie que corresponde a $R_L$.

<div><img src="img/ej1_sintesis_1.jpeg" width="500"/></div>

Como los polos de la expresión ya se encuentran en la posición de los ceros de la transferencia, se pueden remover 2 tanques RC en serie.

<div><img src="img/ej1_sintesis_2.jpeg" width="500"/></div>

Finalmente, como se está exitando el circuito con $I_1$, el primer elemento debe estar en derivación y, por lo tanto, se remueve el último valor constante como una resistencia en derivación.

<div><img src="img/ej1_sintesis_3.jpeg" width="500"/></div>

Siendo el circuito resultante:

![ej1_circuito_resultante](img/ej1_circuito.JPG)

### b) Síntesis analítica

Comenzamos a remover a partir de $Z_{22}$.

$Z_2 = Z_{22} - R_L$

Para $R_L$ se puede elegir cualquier valor que cumpla que $R_L <= 6$ ya que $6$ es lo máximo que se puede remover. Suponiendo $R_L = 4$

$Z_2 = 6. \frac{s^2+8s+12}{s^2+5s+4} - 4 = \frac{6.(s^2+8s+12)- 4.(s^2+5s+4)}{s^2+5s+4} = \frac{6.s^2+48.s+72-4.s^2-20.s-16}{s^2+5s+4} = \frac{2.s^2+28.s+56}{s^2+5s+4}$

Para remover los tanques planteamos

$Z_3 = Z_2 - \frac{K_1}{s+1}$

$K_1 = \lim_{s \to (-1)} Z_2 . (s+1) = \lim_{s \to (-1)} \frac{2.s^2+28.s+56}{(s+1).(s+4)} . (s+1) = 10$

El primer tanque está compuesto por los elementos:

$C_2 = \frac{1}{K_1} = \frac{1}{10} = 0,1f$

$R_3 = \frac{K_1}{\sigma_1} = \frac{10}{1} = 10\Omega$

$Z_3 = Z_2 - \frac{10}{s+1} = \frac{2.s^2+28.s+56}{s^2+5s+4} - \frac{10}{s+1} = \frac{2.s^2+28.s+56-10.(s+4)}{s^2+5s+4} = \frac{2.s^2+28.s+56-10.s-40}{s^2+5s+4} = \frac{2.s^2+18.s+16}{s^2+5s+4}$

$Z_4 = Z_3 - \frac{K_2}{s+4}$

$K_2 = \lim_{s \to (-4)} Z_3 . (s+4) = \lim_{s \to (-4)} \frac{2.s^2+18.s+16}{(s+1).(s+4)} . (s+4) = 8$

El segundo tanque está compuesto por los elementos:

$C_1 = \frac{1}{K_2} = \frac{1}{8} = 0,125f$

$R_2 = \frac{K_2}{\sigma_2} = \frac{8}{4} = 2\Omega$

Finalmente:

$Z_4 = \frac{2.s^2+18.s+16}{s^2+5s+4} - \frac{8}{s+4} =  \frac{2.s^2+18.s+16-8.(s+1)}{s^2+5s+4} = \frac{2.s^2+18.s+16-8.s-8}{s^2+5s+4} = \frac{2.s^2+10.s+8}{s^2+5s+4} = 2.\frac{s^2+5s+4}{s^2+5s+4} = 2$

## Ejercicio 2

Dada la siguiente transferencia de impedancia:

![ej2_consigna](img/ej2_consigna.JPG)

$T(s) = \frac{V_2}{I_1} = \frac{k.(s^2+9)}{s^3+2.s^2+2.s+1}$

**a)** Sintetizar el cuadripolo pasivo sin pérdidas, que cumpla con la **transimpedancia** indicada, cargado a la salida con una impedancia como la que se muestra en la figura.

**b)** Verificar la transimpedancia del circuito obtenido y hallar el valor de **k**.

## Resolución

### a) Sintesis del circuito

Sabiendo que:

$V_2 = Z_{21} . I_1 + Z_{22} . I_2$

$I_2 = -\frac{V_2}{R_L}$

Se puede llegar a que:

$\frac{V_2}{I_1} = \frac{Z_{21}}{1+\frac{Z_{22}}{R_L}}$

Al separar la parte PAR de la IMPAR en el denominador de la transferencia, queda:

$T(s) = \frac{V_2}{I_1} = \frac{k.(s^2+9)}{(s^3+2.s)+(2.s^2+1)}$

Como $s^2+9$ es par, buscamos dividir por la parte impar del denominador:

$T(s) = \frac{\frac{k.(s^2+9)}{s^3+2.s}}{1+\frac{2.s^2+1}{s^3+2.s}}$

Considerando una $R_L$ normalizada, queda entonces que:

$Z_{21}=\frac{k.(s^2+9)}{s^3+2.s}$

$Z_{22}=\frac{2.s^2+1}{s^3+2.s}$

Ahora buscamos sintetizar $Z_{22}$ respetando los ceros de $Z_{21}$, que son en $(-3)$ y en $\infty$.

Como se está excitando el circuito con $I_1$ el primer elemento (de izquierda a derecha) debe estar en derivación.

Como se está sintetizando $Z_{22}$, se remueve comenzando por la salida y finalizando en la entrada.

<div><img src="img/ej2_sintesis_1.jpeg" width="500"/></div>

* Como el polinomio de mayor grado es de orden 3, se puede anticipar que habrá 3 remociones.
* Como se está excitando el circuito con un generador de corriente, el primer elemento (respecto de la entrada) tiene que estar en derivación.

Teniendo en consideración estas condiciones, invierto para que el primer elemento a remover esté en derivación.

Se remueve parcialmente un polo en infinito para desplazar el cero que necesitamos remover en $-3$. Esto implica un capacitor en derivación.

<div><img src="img/ej2_sintesis_2.jpeg" width="500"/></div>

Se invierte y se remueve el polo. Esto implica un tanque LC en serie.

<div><img src="img/ej2_sintesis_3.jpeg" width="500"/></div>

Finalmente se vuelve a invertir para remover un polo en infinito, ya que el primer elemento desde la entrada debe estar en derivación. Esto implica un capacitor en derivación.

<div><img src="img/ej2_sintesis_4.jpeg" width="500"/></div>

Siendo el circuito resultante:

![ej2_circuito_resultante](img/ej2_circuito.JPG)

### b) Verificación

$Z_{22}=\frac{2.s^2+1}{s^3+2.s}$

Invierto:

$Y_1 = \frac{1}{Z_{22}} = \frac{s^3+2.s}{2.s^2+1}$

$Y_2 = Y_1 - K_1 . s$

$K_1 = {Y_1}_{s=j3} . \frac{1}{s} = \frac{s.(s^2+2)}{2.s^2+1} . \frac{1}{s} = \frac{(j3)^2+2}{2.(j3)^2+1} = \frac{-9+2}{2.(-9)+1} = \frac{7}{17}$

Remuevo un capacitor en derivación:

$C_3 = K_1 = \frac{7}{17}$

$Y_2 = \frac{s^3+2.s}{2.s^2+1} - \frac{7}{17} . s = \frac{s^3+2.s-(\frac{7}{17} . s).(2.s^2+1)}{2.s^2+1} = \frac{s^3+2.s - \frac{14}{17} . s^3 - \frac{7}{17} . s}{2.s^2+1} = \frac{\frac{3}{17}.s^3 + \frac{27}{17} . s}{2.s^2+1}$

Invierto:

$Z_2 = \frac{1}{Y_2} = \frac{2.s^2+1}{\frac{3}{17}.s^3 + \frac{27}{17} . s}$

$Z_3 = Z_2 - \frac{2.K_2.s}{s^2+9}$

$2.K_2 = \lim_{s \to (j3)} Z_2 . \frac{s^2+9}{s} = \lim_{s \to (j3)} \frac{2.s^2+1}{\frac{3}{17}.s (s^2 + 9)}. \frac{s^2+9}{s} = \frac{2.(-9)+1}{\frac{3}{17}.(-9)} = \frac{289}{27}$

Remuevo un tanque $LC$ en serie:

$L_1=\frac{2.K_2}{\omega^2} = \frac{289}{243}$

$C_2=\frac{1}{2.K_2} = \frac{27}{289}$

$Z_3 = \frac{2.s^2+1}{\frac{3}{17}.s^3 + \frac{27}{17} . s} - \frac{\frac{289}{27}.s}{s^2+9} = \frac{2.s^2+1 - \frac{289}{27}.s.(\frac{3}{17}.s)}{\frac{3}{17}.s.(s^2 + 9)} = \frac{2.s^2+1-\frac{17}{9}.s^2}{\frac{3}{17}.s.(s^2+9)} = \frac{\frac{1}{9}.s^2+1}{\frac{3}{17}.s.(s^2+9)} = \frac{\frac{1}{9}.(s^2+9)}{\frac{3}{17}.s.(s^2+9)} = \frac{\frac{1}{9}}{\frac{3}{17}.s} = \frac{1}{\frac{27}{17}.s}$

$Z_3$ corresponde a la impedancia de un capacitor que también puede verse como la admitancia:

$Y_3 = \frac{27}{17}.s$

$C_1 = \frac{27}{17}$

Para verificar el valor de la transferencia, puedo utilizar la matriz MAI. Para eso, identifico el circuito de la siguiente manera:

![circuito_con_nodos](img/ej2_circuito_nodos.JPG)

Cortocircuito los nodos 1 y 2. Inyecto corriente $I_0$ y mido $V_0$

$Y_0^0 = s.C_1 + s.C_2 + \frac{1}{s.L_1}$

Cortocircuito los nodos 0 y 2. Inyecto corriente $I_1$ y mido $V_1$

$Y_1^1 = s.C_2 + s.C_3 + \frac{1}{R_L} + \frac{1}{s.L_1}$

Cortocircuito los nodos 0 y 1. Inyecto corriente $I_2$ y mido $V_2$

$Y_2^2 = s.C_1 + s.C_3 + \frac{1}{R_L}$

Se puede armar la matriz MAI como:

$Y_{MAI} = \begin{bmatrix} 
s.C_1 + s.C_2 + \frac{1}{s.L_1} & -s.C_2 - \frac{1}{s.L_1} & -s.C_1 \\ 
-s.C_2-\frac{1}{s.L_1} & s.C_2 + s.C_3 + \frac{1}{R_L} + \frac{1}{s.L_1} & -s.C_3-\frac{1}{R_L} \\ 
-s.C_1 & -s.C_3-\frac{1}{R_L} & s.C_1 + s.C_3 + \frac{1}{R_L}
\end{bmatrix}$

Para averiguar la transimpedancia, se plantea:

$T(s) = \frac{V_2}{I_1} = \frac{V_{12}}{I_{02}} = \frac{Y^{02}_{12}}{Y^0_0}$

$Y^{02}_{12} = -s.C_2-\frac{1}{s.L_1} = -\frac{s^2.C_2.L_1}{s.L_1}$

$Y^0_0 = (s.(C_2 + C_3) + \frac{1}{R_L} + \frac{1}{s.L_1}).(s.(C_1 + C_3) + \frac{1}{R_L})-(-s.C_3-\frac{1}{R_L})^2$

$Y^0_0 = (s.(C_2 + C_3) + \frac{1}{R_L} + \frac{1}{s.L_1}).(s.(C_1 + C_3) + \frac{1}{R_L})-(s.C_3+\frac{1}{R_L})^2$

$Y^0_0 = s^2.(C_2 + C_3).(C_1 + C_3)+s.\frac{C_2 + C_3}{R_L}+s.\frac{C_1 + C_3}{R_L}+\frac{1}{R_L^2}+\frac{C_1 + C_3}{L_1}+\frac{1}{s.L_1.R_L}-s^2.C_3^2-s.\frac{2.C_3}{R_L}-\frac{1}{R_L^2}$

$Y^0_0 = s^2.(C_2.C_1+C_2.C_3+C_3.C_1+C_3^2)+s.\frac{C_1 + C_2 + 2.C_3}{R_L}+\frac{C_1 + C_3}{L_1}+\frac{1}{s.L_1.R_L}-s^2.C_3^2-s.\frac{2.C_3}{R_L}$

$Y^0_0 = s^2.(C_2.C_1+C_2.C_3+C_3.C_1)+s.\frac{C_1 + C_2}{R_L}+\frac{C_1 + C_3}{L_1}+\frac{1}{s.L_1.R_L}$

$Y^0_0 = \frac{s^3.(C_2.C_1+C_2.C_3+C_3.C_1).L_1.R_L+s^2.(C_1 + C_2).L_1+s.(C_1 + C_3).R_L+1}{s.L_1.R_L}$

Finalmente:

$T(s) = \frac{Y^{02}_{12}}{Y^0_0} = \frac{-\frac{s^2.C_2.L_1}{s.L_1}}{\frac{s^3.(C_2.C_1+C_2.C_3+C_3.C_1).L_1.R_L+s^2.(C_1 + C_2).L_1+s.(C_1 + C_3).R_L+1}{s.L_1.R_L}}$

$T(s) = \frac{-R_L.(s^2.C_2.L_1+1)}{s^3.(C_2.C_1+C_2.C_3+C_3.C_1).L_1.R_L+s^2.(C_1 + C_2).L_1+s.(C_1 + C_3).R_L+1}$

$T(s) = \frac{-R_L.C_2.L_1.(s^2+\frac{1}{C_2.L_1})}{s^3.(C_2.C_1+C_2.C_3+C_3.C_1).L_1.R_L+s^2.(C_1 + C_2).L_1+s.(C_1 + C_3).R_L+1}$


```python
#
#            k.(s^2+A)
# T(s) = -----------------
#        s^3.B+s^2.C+s.D+E
#

C1 = 27/17
C2 = 27/289
C3 = 7/17
L1 = 289/243
RL = 1

A = 1/(C2*L1)
B = (C2*C1+C2*C3+C3*C1)*RL*L1
C = (C1+C2)*L1
D = (C1+C3)*RL
k = (-1)*RL*C2*L1

print(f'A: {A}')
print(f'B: {B}')
print(f'C: {C}')
print(f'D: {D}')
print(f'k: {k}')
```

Considerando los coeficientes de la expresión de la **transimpedancia** y los coeficientes calculados con los valores hallados de los componentes, el desarrollo verifica y el valor de la constante es $k=-0,11$
