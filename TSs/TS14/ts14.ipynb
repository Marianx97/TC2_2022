### Resolución

Considerando que es una red pasiva y simétrica, se pueden obtener los parámetros S como:

* $S_{11} = S_{22} = \frac{Z_1-R_0}{Z_1+R_0}$
* $S_{12} = S_{21} = \frac{V_2}{\frac{V_g}{2}} . \sqrt{\frac{R_{01}}{R_{02}}}$

Para poder obtener los parámetros se debe conectar la red:

![red_cargada](img/red_cargada.JPG)

Normalizando tal que:

$R_{01} = R_{02} = R_0$

$R_g = R_L = R_0 = 1$

Entonces:

$S_{11} = \frac{Z_1-1}{Z_1+1}$

$S_{12} = \frac{V_2}{\frac{V_g}{2}}$

La impedancia $Z_1$ se puede expresar como:

$Z_1 = s.L_1 + \frac{1}{s.C_1+\frac{1}{s.L_2+R_L}}$

$Z_1 = s.L_1 + \frac{1}{\frac{s.C_1.(s.L_2+R_L)+1}{s.L_2+R_L}}$

$Z_1 = s.L_1 + \frac{s.L_2+R_L}{s^2.C_1.L_2+s.C_1.R_L+1}$

$Z_1 = \frac{s.L_1.(s^2.C_1.L_2+s.C_1.R_L+1)+s.L_2+R_L}{s^2.C_1.L_2+s.C_1.R_L+1}$

$Z_1 = \frac{s^3.C_1.L_1.L_2+s^2.L_1.C_1.R_L+s.L_1+s.L_2+R_L}{s^2.C_1.L_2+s.C_1.R_L+1}$

$Z_1 = \frac{s^3.C_1.L_1.L_2+s^2.L_1.C_1.R_L+s.(L_1+L_2)+R_L}{s^2.C_1.L_2+s.C_1.R_L+1}$

Reemplazando por los valores:

$Z_1 = \frac{s^3.2.1.1+s^2.1.2.1+s.(1+1)+1}{s^2.2.1+s.2.1+1}$

$Z_1 = \frac{2.s^3+2.s^2+2.s+1}{2.s^2+2.s+1}$

Para calcular el parámetro $S_{11}$:

$S_{11} = \frac{\frac{2.s^3+2.s^2+2.s+1}{2.s^2+2.s+1}-1}{\frac{2.s^3+2.s^2+2.s+1}{2.s^2+2.s+1}+1}$

$S_{11} = \frac{\frac{2.s^3+2.s^2+2.s+1-1.(2.s^2+2.s+1)}{2.s^2+2.s+1}}{\frac{2.s^3+2.s^2+2.s+1+1.(2.s^2+2.s+1)}{2.s^2+2.s+1}}$

$S_{11} = \frac{2.s^3+2.s^2+2.s+1-2.s^2-2.s-1}{2.s^3+2.s^2+2.s+1+2.s^2+2.s+1}$

$S_{11} = \frac{2.s^3}{2.s^3+4.s^2+4.s+2}$

$S_{11} = \frac{2.s^3}{2.(s^3+2.s^2+2.s+1)}$

$S_{11} = \frac{s^3}{s^3+2.s^2+2.s+1}$

Para calcular el parámetro $S_{21}$:

$S_{21} = \frac{V_2}{\frac{V_g}{2}}$

$V_2 = (-I_2) . R_L$

Planteando un divisor de corriente:

$(-I_2) = I_1 . \frac{\frac{1}{s.C_1}}{\frac{1}{s.C_1}+s.L_2+R_L}$

$(-I_2) = I_1 . \frac{1}{s.C_1.(\frac{1}{s.C_1}+s.L_2+R_L)}$

$(-I_2) = I_1 . \frac{1}{1+s^2.C_1.L_2+s.C_1.R_L}$

$(-I_2) = I_1 . \frac{1}{s^2.C_1.L_2+s.C_1.R_L+1}$

$I_1 = \frac{V_g}{R_g+Z_1}$

$I_1 = \frac{V_g}{R_g+\frac{2.s^3+2.s^2+2.s+1}{2.s^2+2.s+1}}$

$I_1 = \frac{V_g}{\frac{R_g.(2.s^2+2.s+1)+2.s^3+2.s^2+2.s+1}{2.s^2+2.s+1}}$

$I_1 = \frac{V_g.(2.s^2+2.s+1)}{R_g.(2.s^2+2.s+1)+2.s^3+2.s^2+2.s+1}$

Recordando que $R_g=1$:

$I_1 = \frac{V_g.(2.s^2+2.s+1)}{2.s^2+2.s+1+2.s^3+2.s^2+2.s+1}$

$I_1 = \frac{V_g.(2.s^2+2.s+1)}{2.s^3+4.s^2+4.s+2}$

Reemplazando las ecuaciones:

$\frac{V_2}{R_L} = \frac{V_g.(2.s^2+2.s+1)}{2.s^3+4.s^2+4.s+2} . \frac{1}{s^2.C_1.L_2+s.C_1.R_L+1}$

Reemplazando por los valores de los componentes:

$\frac{V_2}{1} = \frac{V_g.(2.s^2+2.s+1)}{2.s^3+4.s^2+4.s+2} . \frac{1}{s^2.2.1+s.2.1+1}$

$V_2 = \frac{V_g.(2.s^2+2.s+1)}{2.s^3+4.s^2+4.s+2} . \frac{1}{2.s^2+2.s+1}$

$V_2 = \frac{V_g}{2.s^3+4.s^2+4.s+2}$

$\frac{V_2}{V_g} = \frac{1}{2.s^3+4.s^2+4.s+2}$

$2.\frac{V_2}{V_g} = \frac{2}{2.s^3+4.s^2+4.s+2}$

$2.\frac{V_2}{V_g} = \frac{2}{2.(s^3+2.s^2+2.s+1)}$

$S_{21} = \frac{V_2}{\frac{V_g}{2}} = \frac{1}{s^3+2.s^2+2.s+1}$

Finalmente, la matriz de parámetros S queda:

$S = \begin{bmatrix} S_{11} & S_{12} \\ S_{21} & S_{22}\end{bmatrix}$

$S = \begin{bmatrix} \frac{s^3}{s^3+2.s^2+2.s+1} & \frac{1}{s^3+2.s^2+2.s+1} \\ \frac{1}{s^3+2.s^2+2.s+1} & \frac{s^3}{s^3+2.s^2+2.s+1}\end{bmatrix}$

**a)** Considerando que el parámetro $S_{21}$ describe la transferencia directa de potencia de la red, se puede observar que tiene comportamiento pasa bajos.

**b)** Analizando el comportamiento de los parámetros en función de $\omega$

$S_{11} = \frac{s^3}{s^3+2.s^2+2.s+1}$

${S_{11}}_{(s=j\omega)} = \frac{(j\omega)^3}{(j\omega)^3+2.(j\omega)^2+2.(j\omega)+1}$

${S_{11}}_{(s=j\omega)} = \frac{-j\omega^3}{-j\omega^3-2.\omega^2+j.2.\omega+1}$

${S_{11}}_{(s=j\omega)} = \frac{-j\omega^3}{(1-2.\omega^2)+j(2.\omega-\omega^3)}$

$|{S_{11}}_{(s=j\omega)}| = \frac{|-j\omega^3|}{|(1-2.\omega^2)+j(2.\omega-\omega^3)|}$

$|{S_{11}}_{(s=j\omega)}| = \frac{\sqrt{(\omega^3)^2}}{\sqrt{(1-2.\omega^2)^2+(2.\omega-\omega^3)^2}}$

$S_{21} = \frac{1}{s^3+2.s^2+2.s+1}$

${S_{21}}_{(s=j\omega)} = \frac{1}{(j\omega)^3+2.(j\omega)^2+2.(j\omega)+1}$

${S_{21}}_{(s=j\omega)} = \frac{1}{-j\omega^3-2.\omega^2+j.2.\omega+1}$

${S_{21}}_{(s=j\omega)} = \frac{1}{(1-2.\omega^2)+j(2.\omega-\omega^3)}$

$|{S_{21}}_{(s=j\omega)}| = \frac{1}{|(1-2.\omega^2)+j(2.\omega-\omega^3)|}$

$|{S_{21}}_{(s=j\omega)}| = \frac{1}{\sqrt{(1-2.\omega^2)^2+(2.\omega-\omega^3)^2}}$



Para el centro de la banda de paso: $\omega=0$

$|S_{11}| = 0$

$|S_{21}| = 1$

Para la frecuencia de corte: $\omega=1$

$|S_{11}| = \frac{1}{\sqrt{2}}$

$|S_{21}| = \frac{1}{\sqrt{2}}$

Para el centro de la banda de detención: $\omega \to \infty$

$|S_{11}| = 1$

(Los exponentes de mayor orden quedan iguales, y por eso la indeterminación queda igual a $1$)

$|S_{21}| = 0$

El parámetro $|S_{11}|$ se corresponde con el coeficiente de reflexión. Su comportamiento pasa altos indica que la reflexión será prácticamente nula para bajas frecuencias y total para fecuencias elevadas.

El parámetro $|S_{21}|$ se corresponde con el coeficiente de transferencia de potencia directa. Su comportamiento pasa bajos indica que la transferencia será total para bajas frecuencias y practicamente nula para fecuencias elevadas.

```python
import matplotlib.pyplot as plt
import scipy.signal as sig
from splane import analyze_sys

# COMPORTAMIENTO DEL PARÁMETRO S11: PASA ALTOS
# DESCRIBE EL COMPORTAMIENTO DE LA REFLEXIÓN

num = [1,0,0,0]
den = [1,2,2,1]

transfer_f = sig.TransferFunction(num, den)

plt.close('all')
analyze_sys(transfer_f, '$S_{11}$')
```

```python
# COMPORTAMIENTO DEL PARÁMETRO S21: PASA BAJOS
# DESCRIBE LA TRANSFERENCIA DE POTENCIA

num = [1]
den = [1,2,2,1]

transfer_f = sig.TransferFunction(num, den)

plt.close('all')
analyze_sys(transfer_f, '$S_{21}$')
```
