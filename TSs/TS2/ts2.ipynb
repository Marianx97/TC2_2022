## TS2

Para  la siguiente red se pide:

!["circuito_consigna"](img/circuito_consigna.JPG)

* Hallar la transferencia $T(s) = \frac{V_o}{V_i}$ en función de $\omega_0$ y $Q$
* Obtener el valor de los componentes del circuito de forma tal que $\omega_0 = 1$ y $Q = 3$
* Ajustar el valor de $R_1$ de forma tal que $|T(0)| = 20dB$

## Resolución

Planteo la resolución por nodos:

!["circuito_nodos"](img/circuito_nodos.jpg)

* (1): $0 . (G_1 + G_2 + G_3 + s . C) - V_i . G_1 - V_1 . (G_2 + s . C) - V_o . G_3 = 0$
* (2): $0 . (G_3 + s . C) - V_1 . G_3 - V_2 . s . C = 0$
* (3): $0 . (G_4 + G_4) - V_2 . G_4 - V_o . G_4 = 0$

Partiendo de (3):

(4): $V_2 = -V_o$

Reemplazando (4) en (2):

$V_o . s . C = V_1 . G_3$

(5): $V_1 = V_o . \frac{s . C}{G_3}$

Reemplazando (5) en (1):

$- V_i . G_1 - V_o . \frac{s . C}{G_3} . (G_2 + s . C) - V_o . G_3 = 0$

$- V_i . G_1 = V_o . [\frac{s . C}{G_3} . (G_2 + s . C) + G_3]$

$- V_i . G_1 = V_o . [\frac{s^2.C^2+s.G_2.C+{G_3}^2}{G_3}]$

$T(s) = \frac{V_o}{V_i} = - G_1 . G_3 . \frac{1}{s^2.C^2+s.G_2.C+{G_3}^2}$

$T(s) = \frac{V_o}{V_i} = - \frac{G_1 . G_3}{C^2} . \frac{1}{s^2+s.\frac{G_2}{C}+\frac{{G_3}^2}{C^2}}$

$T(s) = \frac{V_o}{V_i} = - \frac{1}{R_1.R_3.C^2} . \frac{1}{s^2+s.\frac{1}{R_2.C}+\frac{1}{{R_3}^2.C^2}}$

$T(s) = \frac{V_o}{V_i} = - \frac{R_3}{R_1} . \frac{\frac{1}{{R_3}^2.C^2}}{s^2+s.\frac{1}{R_2.C}+\frac{1}{{R_3}^2.C^2}}$

$\omega_0^2 = \frac{1}{{R_3}^2.C^2}$

$\frac{\omega_0}{Q} = \frac{1}{R_2.C}$

### Sensibilidad

$\omega_0 = \frac{1}{R_3.C}$

$Q = \omega_0 . R_2 . C = \frac{1}{R_3 . C} . R_2 . C = \frac{R_2}{R_3}$

$\frac{\Delta\omega_0}{\omega_0} = \frac{{\Delta}C}{C} . S_C^{\omega_0}$

$S_C^{\omega_0} = \frac{C}{\omega_0} . \frac{\delta\omega_0}{{\delta}C}$

$\frac{\delta\omega_0}{{\delta}C} = \frac{1}{R_3} . (- \frac{1}{C^2}) = - \frac{1}{R_3.C^2}$

$S_C^{\omega_0} = \frac{C}{\frac{1}{R_3.C}} . (- \frac{1}{R_3.C^2})$

$S_C^{\omega_0} = R_3 . C^2 . (- \frac{1}{R_3.C^2}) = -1$

$\frac{{\Delta}Q}{Q} = \frac{{\Delta}R_2}{R_2} . S_{R_2}^Q$

$S_{R_2}^Q = \frac{R_2}{Q} . \frac{{\delta}Q}{{\delta}R_2}$

$\frac{{\delta}Q}{{\delta}R_2} = \frac{1}{R_3}$

$S_{R_2}^Q = \frac{R_2}{\frac{R_2}{R_3}} . \frac{1}{R_3} = 1$

$\frac{{\Delta}Q}{Q} = \frac{{\Delta}R_3}{R_3} . S_{R_3}^Q$

$S_{R_3}^Q = \frac{R_3}{Q} . \frac{{\delta}Q}{{\delta}R_3}$

$\frac{{\delta}Q}{{\delta}R_3} = R_2 . (- \frac{1}{{R_3}^2}) = - \frac{R_2}{{R_3}^2}$

$S_{R_3}^Q = \frac{R_3}{\frac{R_2}{R_3}} . (- \frac{R_2}{{R_3}^2}) = - \frac{{R_3}^2}{{R_3}^2} = -1$

Considerando que $\omega_0 = 1$ y $Q = 3$

$1 = \frac{1}{R_3.C}$

$3 = R_2.C$

$k$ representa la ganancia:

$k = \frac{R_3}{R_1}$

$R_1 = \frac{R_3}{k}$

Para tener $20dB$ en continua como pide la consigna:

$20db = 20 . log(k)$

$k = 10$

Suponiendo $C = 1$:

$R_1 = 0,1$

$R_2 = 3$

$R_3 = 1$

```python
import numpy as np

# Componentes normalizados
R1n = 0.1
R2n = 3
R3n = 1
Cn = 1

# Norma de frecuencia
Ww = 2*np.pi*1000

# Norma de impedancia
Wz = 1000

# Componentes desnormalizados
R1dn = R1n*Wz
R2dn = R2n*Wz
R3dn = R3n*Wz
Cdn = Cn/(Ww*Wz)

print(f'R1dn: {R1dn}')
print(f'R2dn: {R2dn}')
print(f'R3dn: {R3dn}')
print(f'Cdn: {Cdn}')
```

## Transferencia Butterworth

Transferencia butterworth pasa-bajos:

$T(s) = \frac{1}{s^2 + s . \sqrt{2} + 1}$

$\omega_0^2 = \frac{1}{{R_3}^2.C^2} = 1$

$\frac{\omega_0}{Q} = \frac{1}{R_2.C} = \sqrt{2}$

$k = \frac{R_3}{R_1} = 1$

Supongo $C = 1$

$R_1 = 1$

$R_2 = \frac{1}{\sqrt{2}}$

$R_3 = 1$

```python
# Simulacion numérica - Butterworth

import matplotlib.pyplot as plt
import scipy.signal as sig
import numpy as np
from splane import analyze_sys

num = [1]
den = [1, np.sqrt(2), 1]

transfer_f = sig.TransferFunction(num, den)

plt.close('all')

analyze_sys(transfer_f, 'Transferencia')
```

```python
# Componentes normalizados
R1n = 1
R2n = 1/np.sqrt(2)
R3n = 1
Cn = 1

# Norma de frecuencia
Ww = 2*np.pi*1000

# Norma de impedancia
Wz = 1000

# Componentes desnormalizados
R1dn = R1n*Wz
R2dn = R2n*Wz
R3dn = R3n*Wz
Cdn = Cn/(Ww*Wz)

print(f'R1dn: {R1dn}')
print(f'R2dn: {R2dn}')
print(f'R3dn: {R3dn}')
print(f'Cdn: {Cdn}')
```

## Transferencia Pasa-Banda

Utilizando el mismo circuito, podemos obtener un pasa-banda tomando la salida del primer operacional

!["circuito_nodos_pasa_banda"](img/circuito_nodos_pb.jpg)

* (1): $- V_i . G_1 - V_o . (G_2 + s . C) - V_1 . G_3 = 0$
* (2): $- V_o . G_3 - V_2 . s . C = 0$
* (3): $- V_2 . G_4 - V_1 . G_4 = 0$

Partiendo de (3):

(4): $V_1 = - V_2$

Partiendo de (2):

(5): $V_2 = -\frac{V_o.G_3}{s.C}$

Considerando (4) y (5):

(6): $V_1 = \frac{V_o.G_3}{s.C}$

Reemplazando (6) en (1):

$- V_i . G_1 - V_o . (G_2 + s . C) - V_o . \frac{G_3}{s.C} . G_3 = 0$

$- V_i . G_1 - V_o . [(G_2 + s . C) + . \frac{{G_3}^2}{s.C}]= 0$

$-V_i . G_1 = V_o . (\frac{s^2.C^2+s.G_2.C+{G_3}^2}{s.C})$

$T(s) = \frac{V_o}{V_i} = - \frac{s.G_1.C}{s^2.C^2+s.G_2.C+{G_3}^2}$

$T(s) = - \frac{G_1.C}{C^2} . \frac{s}{s^2+s.\frac{G_2}{C}+\frac{{G_3}^2}{C^2}}$

$T(s) = - \frac{1}{R_1.C} . \frac{s}{s^2+s.\frac{1}{R_2.C}+\frac{1}{{R_3}^2.C^2}}$

Utilizando los mismos componentes que en el primer caso:

$R_1 = 0,1$

$R_2 = 3$

$R_3 = 1$

$C = 1$

$\omega_0^2 = 1$

$\frac{\omega_0}{Q} = \frac{1}{R_2.C} = \frac{1}{3}$

$Q = 3$

$\frac{1}{R_1.C} = k . \frac{1}{R_2.C}$

$k = \frac{R_2}{R_1} = 30$

```python
# Simulacion numérica

import matplotlib.pyplot as plt
import scipy.signal as sig
import numpy as np
from splane import analyze_sys

num = [-10,0]
den = [1,(1/3),1]

transfer_f = sig.TransferFunction(num, den)

plt.close('all')

analyze_sys(transfer_f, 'Transferencia')
```
