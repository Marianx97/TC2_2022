Para obtener la aproximación de Bessel de la función de transferencia del filtro pasa bajos de orden 3 utilizamos la siguiente tabla obtenida del material de estudio:

![tabla_bessel](img/tabla_bessel.JPG)

Recordando que:

$T(s) = \frac{K_0}{Be_n(s)}$

Entonces:

$T(s) = \frac{15}{s^3+6.s^2+15.s+15}$

Buscamos hallar la expresión de $|T(s)|^2$

$|T(s)|^2 = T(s) . T(-s)$

$T(s) = \frac{15}{s^3+6.s^2+15.s+15}$

$T(-s) = \frac{15}{-s^3+6.s^2-15.s+15}$

Entonces:

$|T(s)|^2 = \frac{15}{s^3+6.s^2+15.s+15} . \frac{15}{-s^3+6.s^2-15.s+15}$

$|T(s)|^2 = \frac{225}{-s^6 + 6.s^5 - 15.s^4 + 15.s^3 -6.s^5 + 36.s^4 - 90.s^3 + 90.s^2 -15.s^4 + 90.s^3 - 225.s^2 + 225.s -15.s^3 + 90.s^2 - 225.s + 225}$

$|T(s)|^2 = \frac{225}{-s^6+6.s^4-45.s^2+225}$

$|T(s)|^2 = |S_{21}|^2$

Por lo tanto:

$T(s) = S_{21} = \frac{15}{s^3+6.s^2+15.s+15}$

$|S_{11}|^2 + |S_{21}|^2 = 1$

$|S_{11}|^2 = 1 - |S_{21}|^2$

$|S_{11}|^2 = 1 - \frac{225}{-s^6+6.s^4-45.s^2+225}$

$|S_{11}|^2 = \frac{(-s^6+6.s^4-45.s^2+225) - 225}{-s^6+6.s^4-45.s^2+225}$

$|S_{11}|^2 = \frac{(-1)}{(-1)} \frac{s^6-6.s^4+45.s^2-225+225}{s^6-6.s^4+45.s^2-225}$

$|S_{11}|^2 = \frac{s^6-6.s^4+45.s^2}{s^6-6.s^4+45.s^2-225}$

Para $S_{11}$ sabemos que el denominador va a ser: $s^3+6.s^2+15.s+15$. Falta hallar el numerador.

```python
# UTILIZAMOS PYTHON PARA HALLAR LAS RAÍCES DEL NUMERADOR
import numpy as np

num = [1, 0, -6, 0, 45, 0, 0]

num_roots = np.roots(num)

# Nos quedamos con las raices cuya parte real es negativa para armar el sistema
for root in num_roots:
    if root.real <= 0:
        print(f'raiz num: {root}')
```

Operando con las raíces obtenidas, queda:

$S_{11} = \frac{s^3+4,4.s^2+6,69.s}{s^3+6.s^2+15.s+15}$

**1.** Para obtener la impedancia de entrada del cuadripolo A despejamos $Z_1$ de la siguiente expresión:

$S_{11} = \frac{Z_1-R_0}{Z_1+R_0}$

$S_{11} . (Z_1+R_0) = Z_1-R_0$

$S_{11} . Z_1 + S_{11} . R_0 = Z_1 - R_0$

$S_{11} . R_0 + R_0 = Z_1 - S_{11} . Z_1$

$(S_{11} + 1) . R_0 = (1 - S_{11}) . Z_1$

$Z_1 = R_0 . \frac{S_{11} + 1}{1 - S_{11}}$

Reemplazando con la expresión obtenida para $S_{11}$ y recordando que $R_0=1$:

$Z_1 = \frac{\frac{s^3+4,4.s^2+6,69.s}{s^3+6.s^2+15.s+15} + 1}{1 - \frac{s^3+4,4.s^2+6,69.s}{s^3+6.s^2+15.s+15}}$

$Z_1 = \frac{\frac{s^3+4,4.s^2+6,69.s+s^3+6.s^2+15.s+15}{s^3+6.s^2+15.s+15}}{\frac{s^3+6.s^2+15.s+15-(s^3+4,4.s^2+6,69.s)}{s^3+6.s^2+15.s+15}}$

$Z_1 = \frac{s^3+4,4.s^2+6,69.s+s^3+6.s^2+15.s+15}{s^3+6.s^2+15.s+15-s^3-4,4.s^2-6,69.s}$

$Z_1 = \frac{2.s^3+10,4.s^2+21,69.s+15}{1,6.s^2+8,31.s+15}$
