## TS12

### Ejercicio 1

Sintetizar un cuadripolo que cumpla con los siguientes parámetros:

$Y_{11} = \frac{I_1}{V_1}|_{V_2=0} = \frac{3s.(s^2+\frac{7}{3})}{(s^2+2).(s^2+5)}$

$Y_{21} = \frac{I_2}{V_1}|_{V_2=0} = \frac{s.(s^2+1)}{(s^2+2).(s^2+5)}$

**a)** Obtener la topología mediante la **síntesis gráfica**, es decir la red sin valores.

**b)** Calcular el valor de los componentes, es decir la **síntesis analítica**.

### Ejercicio 2

Dada la siguiente transferencia:

$T(s) = \frac{V_2}{V_1}|_{I_2=0} = \frac{k.(s+1)}{(s+2).(s+4)}$

**a)** Obtener la topología circuital que respeta la transferencia solicitada, utilizando parámetros Z e Y.

**b)** Calcular el valor de los componentes y el parámetro $k$.

### Ejercicio 1 - Resolución

$Y_{11} = \frac{I_1}{V_1}|_{V_2=0} = \frac{3s.(s^2+\frac{7}{3})}{(s^2+2).(s^2+5)}$

$Y_{21} = \frac{I_2}{V_1}|_{V_2=0} = \frac{s.(s^2+1)}{(s^2+2).(s^2+5)}$

Se debe sintetizar $Y_{11}$ respetando los ceros de $Y_{21}$.

Síntesis gráfica:

<div><img src="img/sintesis_grafica_1.jpeg" width="500"/></div>

Invertimos: $Z_2 = \frac{1}{Y_{11}}$

Se remueve parcialmente en continua para desplazar el cero y que coincida con la posición del cero de $Y_{21}$. La remoción de un polo en el origen implica un capacitor en serie.

<div><img src="img/sintesis_grafica_2.jpeg" width="500"/></div>

Invertimos: $Y_4 = \frac{1}{Z_4}$

Se remueve un polo en $1$ como un tanque serie en derivación.

<div><img src="img/sintesis_grafica_3.jpeg" width="500"/></div>

Invertimos: $Z_6 = \frac{1}{Y_6}$

A partir de este punto se puede remover un polo en cero y un polo en infinito, lo cual implica un capacitor en serie y un inductor en serie.

<div><img src="img/sintesis_grafica_4.jpeg" width="500"/></div>

Finalmente la red queda:

![red_circuital](img/red_circuital.JPG)

Síntesis analítica:

$Y_{11} = \frac{I_1}{V_1}|_{V_2=0} = \frac{3s.(s^2+\frac{7}{3})}{(s^2+2).(s^2+5)}$

$Z_2 = \frac{1}{Y_{11}} = \frac{(s^2+2).(s^2+5)}{3s.(s^2+\frac{7}{3})}$

$Z_4 = Z_2 - \frac{K_0}{s}$

Para respetar los ceros de $Y_{21}$ planteamos el residuo como:

$K_0 = \lim_{s^2 \to (-1)} Z_2.s = \lim_{s^2 \to (-1)} \frac{(s^2+2).(s^2+5)}{3.(s^2+\frac{7}{3})} = \frac{(-1+2).(-1+5)}{3.(-1+\frac{7}{3})} = 1$

Esto implica un capacitor en serie: $C_1 = 1$

$Z_4 = \frac{(s^2+2).(s^2+5)}{3s.(s^2+\frac{7}{3})} - \frac{1}{s} = \frac{(s^2+2).(s^2+5)-3.(s^2+\frac{7}{3})}{3s.(s^2+\frac{7}{3})} = \frac{s^4+7.s^2+10-3.s^2-7}{3s.(s^2+\frac{7}{3})}$

$Z_4 = \frac{s^4+4.s^2+3}{3s.(s^2+\frac{7}{3})} = \frac{(s^2+1).(s^2+3)}{3s.(s^2+\frac{7}{3})}$

Invertimos:

$Y_4 = \frac{1}{Z_4} = \frac{3s.(s^2+\frac{7}{3})}{(s^2+1).(s^2+3)}$

Removemos el polo en la posición del cero de $Y_{21}$

$Y_6 = Y_4 - \frac{K_1.s}{s^2+1}$

$K_1 = \lim_{s^2 \to (-1)} Y_4 . \frac{s^2+1}{s} = \lim_{s^2 \to (-1)} \frac{3.(s^2+\frac{7}{3})}{s^2+3} = \frac{3.(-1+\frac{7}{3})}{-1+3} = 2$

$\frac{K_1.s}{s^2+1} = \frac{2.s}{s^2+1} = \frac{1}{\frac{s}{2}+\frac{1}{2.s}}$

Esto implica un tanque LC serie en derivación donde $L_1 = \frac{1}{2}$ y $C_2 = 2$

$Y_6 = \frac{3s.(s^2+\frac{7}{3})}{(s^2+1).(s^2+3)} - \frac{2.s}{s^2+1} = \frac{3s.(s^2+\frac{7}{3})-2.s.(s^2+3)}{(s^2+1).(s^2+3)} = \frac{3.s^3+7.s-2.s^3-6.s}{(s^2+1).(s^2+3)}$

$Y_6 = \frac{s^3+s}{(s^2+1).(s^2+3)} = \frac{s.(s^2+1)}{(s^2+1).(s^2+3)} = \frac{s}{s^2+3}$

Invertimos:

$Z_6 = \frac{1}{Y_6} = \frac{s^2+3}{s} = s + \frac{1}{\frac{1}{3}.s}$

Esto implica un capacitor y un inductor en serie donde: $L_2 = 1$ y $C_3 = \frac{1}{3}$

### Ejercicio 2 - Resolución

$T(s) = \frac{V_2}{V_1}|_{I_2=0} = \frac{-Y_{21}}{Y_{11}} = \frac{k.(s+1)}{(s+2).(s+4)}$

Debemos sintetizar $Y_{11}$ respetando los ceros de $Y_{21}$ para respetar así los ceros de la transferencia. Para poder hacer esto, proponemos un denominador tal que:

$T(s) = \frac{\frac{-Y_{21}}{D(s)}}{\frac{Y_{11}}{D(s)}}$

El denominador $T(s)$ debe ser tal que la sintesis de $\frac{Y_{11}}{D(s)}$ tenga alternancia de polos y ceros. Planteamos:

$\frac{Y_{11}}{D(s)} = \frac{(s+2).(s+4)}{s+3}$

Síntesis gráfica:

<div><img src="img/sintesis_grafica_ej2_1.jpeg" width="500"/></div>

Realizamos una remoción parcial en continua para desplazar el cero y que coincida con el cero de $Y_{21}$. Esto implica un resistor en derivación.

<div><img src="img/sintesis_grafica_ej2_2.jpeg" width="500"/></div>

Invertimos $Z_2 = \frac{1}{Y_2}$

<div><img src="img/sintesis_grafica_ej2_3.jpeg" width="500"/></div>

Removemos el polo en $-1$. Esto implica un tanque $RC$ paralelo en serie.

<div><img src="img/sintesis_grafica_ej2_4.jpeg" width="500"/></div>

Invertimos $Y_4 = \frac{1}{Z_4}$

<div><img src="img/sintesis_grafica_ej2_5.jpeg" width="500"/></div>

Removemos el polo en infinito. Esto implica un capacitor en derivación. Luego invertimos $Z_6 = \frac{1}{Y_6}$. El resultado es un resistor en serie

<div><img src="img/sintesis_grafica_ej2_6.jpeg" width="500"/></div>

El circuito resultante es:

![ej2_red_circuital](img/ej2_red_circuital.JPG)

Sintesis analítica:

$\frac{Y_{11}}{D(s)} = \frac{(s+2).(s+4)}{s+3}$

$Y_2 = \frac{Y_{11}}{D(s)} - K_1$

Planteamos el residuo $K_1$ como:

$K_1 = \lim_{s \to (-1)} \frac{Y_{11}}{D(s)} = \lim_{s \to (-1)} \frac{(s+2).(s+4)}{s+3} = \frac{(-1+2).(-1+4)}{-1+3} = \frac{3}{2}$

Esto implica un resistor en derivación $R_1 = \frac{3}{2}$

$Y_2 = \frac{(s+2).(s+4)}{s+3} - \frac{3}{2} = \frac{s^2+6.s+8}{s+3} - \frac{3}{2} = \frac{2.s^2+12.s+16-3.s-9}{2.(s+3)} = \frac{2.s^2+9.s+7}{2.(s+3)} = \frac{(s+1).(s+\frac{7}{2})}{s+3}$

Invertimos $Z_2 = \frac{1}{Y_2}$

$Z_2 = \frac{s+3}{(s+1).(s+\frac{7}{2})}$

$Z_4 = Z_2 - \frac{K_2}{s+1}$

Planteamos el residuo $K_2$ como:

$K_2 = \lim_{s \to (-1)} Z_2 . (s+1) = \lim_{s \to (-1)} \frac{s+3}{s+\frac{7}{2}} = \frac{-1+3}{-1+\frac{7}{2}} = \frac{4}{5}$

Determinamos los valores de los componentes del tanque como:

$\frac{K_2}{s+1} = \frac{\frac{4}{5}}{s+1} = \frac{1}{\frac{5.s}{4}+\frac{5}{4}}$

Esta expresión corresponde al paralelo entre $C_1 = \frac{5}{4}$ y $R_2 = \frac{4}{5}$

$Z_4 = \frac{s+3}{(s+1).(s+\frac{7}{2})} - \frac{4}{5.(s+1)}$

$Z_4 = \frac{5.(s+3)-4.(s+\frac{7}{2})}{5.(s+1).(s+\frac{7}{2})} = \frac{5.s+15-4.s-14}{5.(s+1).(s+\frac{7}{2})} = \frac{s+1}{5.(s+1).(s+\frac{7}{2})} = \frac{1}{5.(s+\frac{7}{2})}$

Invertimos $Y_4 = \frac{1}{Z_4} = 5.(s+\frac{7}{2})$

$Y_4 = 5.s+\frac{1}{\frac{2}{35}}$

$Y_6 = Y_4 - K_3 . s$

Planteamos el residuo $K_3$ como:

$K_3 = \lim_{s \to \infty} \frac{Y_4}{s} = \lim_{s \to \infty} \frac{5.(s+\frac{7}{2})}{s} = 5$

Esto implica un capacitor en derivación de valor $C_2 = 5$

$Y_6 = 5.(s+\frac{7}{2}) - 5 . s = 5 . (s+\frac{7}{2}-s) = 5.\frac{7}{2} = \frac{35}{2}$

Invertimos $Z_6 = \frac{1}{Y_6} = \frac{2}{35}$. Esto implica un resistor en serie de valor $R_3 = \frac{2}{35}$

Finalmente la red queda:

![ej2_red_circuital_con_valores](img/ej2_red_circuital_con_valores.JPG)

Para determinar el valor de $k$, planteamos el valor de la transferencia cuando $s \to 0$. En continua los capacitores se comportan como circuitos abiertos, por lo tanto:

$\frac{V_2}{V_1} = \frac{\frac{3}{2}}{\frac{3}{2}+(\frac{2}{35}+\frac{4}{5})} = \frac{7}{11}$

En la expresión matemática:

$\frac{V_2}{V_1} = \lim_{s \to 0} \frac{k.(s+1)}{(s+2).(s+4)} = \frac{k}{8}$

Al igualar ambas expresiones:

$\frac{7}{11} = \frac{k}{8}$

$k = \frac{56}{11}$
